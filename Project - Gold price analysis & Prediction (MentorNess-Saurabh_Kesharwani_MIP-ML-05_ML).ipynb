# Model building

```python
# Model building by fb prophet
```

```python
from prophet import Prophet
```

```python
df
```

```python
df_close= df[['Date','Close']]
df_close
```

```python
df_open= df[['Date','Open']]
df_open
```

```python
# Lets do forecasting for closing price
```

```python
# renaming the columns-- 'ds' and 'y'
```

```python
df_close.rename(columns={'Date':'ds', 'Close':'y'}, inplace=True)
df_close
```

```python
# Model building
```

```python
# Split your data into training and testing sets
train_size = int(0.8 * len(df_close)) 
train_size
```

```python
train_df = df_close[:train_size]
test_df = df_close[train_size:]
```

```python
train_df
```

```python
model=Prophet(interval_width=0.95)
```

```python
model.fit(train_df)
```

```python
# Make predictions on the test data
future = model.make_future_dataframe(periods=len(test_df))
```

```python
future.tail(10)
```

```python
forecast=model.predict(future)
```

```python
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(10)
```

```python
# Extract actual and predicted values
actual_values = test_df['y'].values
predicted_values = forecast['yhat'][-len(test_df):].values
```

```python
actual_values
```

```python
predicted_values
```

# Model Evaluation for Closing price

```python
from sklearn.metrics import mean_squared_error
```

```python
mse_close = mean_squared_error(actual_values, predicted_values)
print("Mean Squared Error (MSE) for closing price:", mse_close)
```

```python
rmse_close = np.sqrt(mse_close)
```

```python
rmse_close
```

```python
df_close_plot=model.plot(forecast)
```

```python
df_close_plot=model.plot_components(forecast)
```

```python
# Lets do forecasting for Opening price
```

```python
df_open.rename(columns={'Date':'ds', 'Open':'y'}, inplace=True)
df_open
```

```python
# Model building  for opening price
```

```python
# Split your data into training and testing sets
train_size = int(0.8 * len(df_open)) 
train_size
```

```python
train_df = df_open[:train_size]
test_df = df_open[train_size:]
```

```python
train_df
```

```python
model2=Prophet(interval_width=0.95)
```

```python
model2.fit(train_df)
```

```python
# Make predictions on the test data
future2 = model2.make_future_dataframe(periods=len(test_df))
```

```python
future2.tail(10)
```

```python
forecast_open=model2.predict(future2)
```

```python
forecast_open[['ds','yhat','yhat_lower','yhat_upper']].tail(10)
```

```python
# Extract actual and predicted values
actual_values = test_df['y'].values
predicted_values = forecast_open['yhat'][-len(test_df):].values
```

```python
actual_values
```

```python
predicted_values
```

# Model Evaluation for Opening price

```python
mse_open = mean_squared_error(actual_values, predicted_values)
print("Mean Squared Error (MSE) for opening price:", mse_open)
```

```python
rmse_open = np.sqrt(mse_open)
```

```python
rmse_open
```

```python
df_open_plot=model2.plot(forecast_open)
```

```python
df_open_plot=model2.plot_components(forecast_open)
```
