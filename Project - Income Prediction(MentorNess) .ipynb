```python
#  Replacing "?" with 'nan' i.e. with null value. 

df_train=df_train.replace('?', np.nan)
```

```python
df_train
```

```python
df_test=df_test.replace('?', np.nan)
```

```python
df_test
```

```python
# now, checking null values
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
# replacing  the null values with mode  -- for train data
# only categorical column has null value
```

```python
for col in df_train.columns:
    if df_train[col].dtype == 'object':
        mode = df_train[col].mode()[0]
        df_train[col] = df_train[col].fillna(mode)
```

```python
df_train
```

```python
# Replacing the null value of test data with mode

for col in df_test.columns:
    if df_test[col].dtype == 'object':
        mode = df_test[col].mode()[0]
        df_test[col] = df_test[col].fillna(mode)
```

```python
df_test
```

```python
# outliers for the test dataset
for col in df_test.columns:
    if df_test[col].dtype=='int64' or df_test[col].dtype=='float64':
        plt.figure(figsize=(5,3))
        sns.boxplot(df_test[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
# only column 'importance_of_record' shows outliers with proper boxplot. so we will remove outitiers from this column
```

```python
# dropping outliers from training dataset
```

```python
Q1=df_train['importance_of_record'].quantile(0.25)
Q3=df_train['importance_of_record'].quantile(0.75)
IQR=Q3-Q1
df_train=df_train[(df_train['importance_of_record']>=Q1-1.5*IQR) & (df_train['importance_of_record']<=Q3+1.5*IQR)]
```

```python
# again check outlier
for col in df_train.columns:
    if df_train[col].dtype=='int64' or df_train[col].dtype=='float64':
        plt.figure(figsize=(5,3))
        sns.boxplot(df_train[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
# dropping outliers from test dataset

Q1=df_test['importance_of_record'].quantile(0.25)
Q3=df_test['importance_of_record'].quantile(0.75)
IQR=Q3-Q1
df_test=df_test[(df_test['importance_of_record']>=Q1-1.5*IQR) & (df_test['importance_of_record']<=Q3+1.5*IQR)]
```

```python
for col in df_test.columns:
    if df_test[col].dtype=='int64' or df_test[col].dtype=='float64':
        plt.figure(figsize=(5,3))
        sns.boxplot(df_test[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```

```python
# features selection for training dataset
```

```python
x=df_train.iloc[:,:-1]
y=df_train.iloc[:,-1]
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model= LogisticRegression()
```

```python
rfe=RFE(model)
```

```python
rfe.fit(x,y)
```

```python
print(rfe.support_)
```

```python
print(rfe.ranking_)
# lowest ranking gives most important features.
```

```python
# index of selected features

selected_indices=[i for i, data  in enumerate(rfe.support_) if data==True]
```

```python
selected_indices
```

```python
column_names=x.columns
```

```python
column_names
```

```python
selected_column_names=[column_names[i] for i in selected_indices]
```

```python
x=x.loc[:,['age','gender','education','education_institute','marital_status','race','is_hispanic', 'employment_commitment', 'working_week_per_year',
           'industry_code_main','occupation_code','occupation_code_main','total_employed','household_summary','under_18_family','vet_benefit',
           'tax_status','mig_year','migration_code_change_in_msa','migration_prev_sunbelt','old_residence_state']]
```

```python
df_test_selected_feature=df_test.loc[:,['age','gender','education','education_institute','marital_status','race','is_hispanic', 'employment_commitment', 'working_week_per_year',
           'industry_code_main','occupation_code','occupation_code_main','total_employed','household_summary','under_18_family','vet_benefit',
           'tax_status','mig_year','migration_code_change_in_msa','migration_prev_sunbelt','old_residence_state']]
```

```python
# splitting the dataset into train and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=11)
```

```python
#checking imbalance of dataset
```

```python
y_train.value_counts()
```

```python
sns.countplot(x=y_train)
plt.title('Imbalanced data')
plt.show()
```

```python
# data is imbalance, individual Income > 50000 is much higher.
```

```python
# to handle the imbalance data, use SMOTE technique
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
```

```python
x_train,y_train=smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
sns.countplot(x=y_train)
plt.title('Balanced data')
plt.show()
```

```python
# now, buildind Logistic Regression Model
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# predict the train data
```

```python
y_pred_train=lr.predict(x_train)
```

```python
# predict test data
```

```python
y_pred_test=lr.predict(x_test)
```

```python
# evaluation of model
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# accuracy of train data
accuracy_score(y_train,y_pred_train)
```

```python
# accuracy of test data
accuracy_score(y_test,y_pred_test)
```

```python
cm=confusion_matrix(y_test,y_pred_test)
cm
```

```python
sns.heatmap(cm,annot=True, cmap="Blues",fmt="d")
```
